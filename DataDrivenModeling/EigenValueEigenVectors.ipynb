```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
oriimg = cv2.imread('zeeshan.jpg',0)
```

```python
newimg = cv2.resize(oriimg,(64,64))
```

```python
plt.imshow(newimg)
```

```python
n=newimg.shape[0]*newimg.shape[1]
x=newimg.reshape(1,n)
```

```python
corrmat=np.matmul(x.T,x)
```

```python
corrmat.shape
```

```python
V, D = np.linalg.eig(corrmat)
```

```python
V.shape
```

```python
D.shape
```

```python
plt.plot(V[0:30])
plt.show()
```

```python
D_r=D[:,0:10]
```

```python
y=np.linspace(1,10,10)
```

```python
proj_x=np.matmul((x),D_r)
```

```python
plt.hist(np.real(proj_x),y)
```
